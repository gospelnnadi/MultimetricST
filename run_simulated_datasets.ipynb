# Evalutaion Simulated dataset using Silhouette spatial coefficient

```python

import scanpy as sc
import numpy as np
import os
from MultimetricST import preprocess, plot_label
from Evaluate.evaluate import evaluate_cluster
plot_size=0.02
adata= sc.read_h5ad("Data/SRTsim-SyntheticData/SRTsim_synthetic2.h5ad")
pca_matrix = adata.obsm["X_pca"]
plot_savepath=f"figures_simulated_data/Simulated_data"
os.makedirs(plot_savepath, exist_ok=True)
print(adata)

method="Case10"
plot_label(adata, plot_size=plot_size, key=method, savepath=f"{plot_savepath}/{method}" )
pred=adata.obs[method]
scores=evaluate_cluster( adata,pred, None, pca_matrix, is_visium=False,verbose=True,decimal=4)

#perm
print("\nPermuting spatial coordinates...\n")

n = adata.obsm['spatial'].shape[0]
rng = np.random.default_rng(seed=43)
perm = rng.choice(n, size=n, replace=False)
adata.obsm['spatial'] = adata.obsm['spatial'][perm]
plot_label(adata, plot_size=plot_size, key=method, savepath=f"{plot_savepath}/{method}_perm" )
pred=adata.obs[method]
scores=evaluate_cluster( adata,pred, None, pca_matrix, is_visium=False,verbose=True,decimal=4)
```
